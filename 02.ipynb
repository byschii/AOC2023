```python
# a game if possible if can be played w -> 12 red cubes, 13 green cubes, and 14 blue cubes

MAX_RED = 12
MAX_GREEN = 13
MAX_BLUE = 14

class Extraction: # from bag
    def __init__(self, n_red, n_green, n_blue):
        self.n_red = n_red
        self.n_green = n_green
        self.n_blue = n_blue

    def is_possible(self, max_red, max_green, max_blue):
        red_ok = self.n_red <= max_red
        green_ok = self.n_green <= max_green
        blue_ok = self.n_blue <= max_blue
        tot_ok = sum([self.n_red,self.n_green,self.n_blue]) <= sum([max_red, max_green, max_blue])
        return red_ok and green_ok and blue_ok and tot_ok
    

    def __repr__(self):
        return f"E[r={self.n_red},g={self.n_green},b={self.n_blue}]"

class Game:
    def __init__(self, id, extractions, txt):
        self.id = id
        self.extractions = extractions
        self.txt = txt

    def is_possible(self, max_red, max_green, max_blue) -> bool:
        return all(list(map(
            lambda e: e.is_possible(max_red, max_green, max_blue), 
            self.extractions
            )))
    
    def __repr__(self):
        return f"G id={self.id},e={self.extractions},txt={self.txt}"
```

```python

games = []
for line in puzzle_input.splitlines():
    if len(line)> 0:
        game_id, extractions_result = line.split(":")
        game_id = int(game_id.split(" ")[1])

        extractions_text = extractions_result.split(";")
        extractions = []
        for e in extractions_text:
            extraction_numbers = {
                "red":0,
                "green":0,
                "blue":0
            }
            for number_color in e.split(","):
                num, color = number_color.strip().split()
                extraction_numbers[color] = int(num)
            extractions.append(Extraction(
                extraction_numbers["red"], extraction_numbers["green"], extraction_numbers["blue"]
            ))

        games.append(Game(game_id, extractions, line))

```

```python
for game in games:
    if game.is_possible(MAX_RED,MAX_GREEN,MAX_BLUE):
        print(game)
```

```python
sum(list(map(lambda g: g.id,
    filter(
    lambda g: g.is_possible(MAX_RED,MAX_GREEN,MAX_BLUE),
    games
    ))))
```

```python
def get_minimal_set_cubes(game):
    mins = {
        "red":0, "green":0, "blue":0
    }

    # get 'color' cubes extracted
    mins["red"] = max(list(map(lambda e: e.n_red, game.extractions)))
    mins["green"] = max(list(map(lambda e: e.n_green, game.extractions)))
    mins["blue"] = max(list(map(lambda e: e.n_blue, game.extractions)))

    return mins

get_minimal_set_cubes(games[-2])

```

```python
game_powers = []
for game in games:
    mins = get_minimal_set_cubes(game)
    game_powers.append(
        mins["red"] * mins["green"] * mins["blue"]
    )

print(sum(game_powers))
```
