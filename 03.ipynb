```python
from collections import Counter
import string 

symbols = list(
    filter(
        lambda c:c not in string.digits + '.',
        Counter(puzzle_input.replace("\n","")).keys()
    ))

symbols
```

```python
# IDEA
# 1. find number
# 2. remember coordinates of digits
# 3. check around every digit

def look_around1(x,y):
    for i in range(-1,2):
        for j in range(-1, 2):
            _x = min(max(x+i,0), len(engine_matrix)-1)
            _y = min(max(y+j,0), len(engine_matrix)-1)
            if engine_matrix[_x][_y] in symbols:
                return True
    return False

def look_around2(x,y):
    # just to remain in the matrix (assuming engine_matrix is square)
    mm = lambda c : min(max(c,0), len(engine_matrix)-1)

    adjacency = [
        engine_matrix[mm(x-1)][mm(y-1)],
        engine_matrix[x][mm(y-1)],
        engine_matrix[mm(x+1)][mm(y-1)],
        engine_matrix[mm(x-1)][y],
        engine_matrix[mm(x+1)][y],
        engine_matrix[mm(x-1)][mm(y+1)],
        engine_matrix[x][mm(y+1)],
        engine_matrix[mm(x+1)][mm(y+1)]
    ]

    return any(map(
        lambda a:a in symbols, adjacency
    ))

def look_around3(x,y):
    # just to remain in the matrix (assuming engine_matrix is square)
    mm = lambda c : min(max(c,0), len(engine_matrix)-1)

    around = {
        (mm(x-1),mm(y-1)),
        (x,mm(y-1)),
        (mm(x+1),mm(y-1)),
        (mm(x-1),y),
        (mm(x+1),y),
        (mm(x-1),mm(y+1)),
        (x,mm(y+1)),
        (mm(x+1),mm(y+1))
    }

    adjacency = [
        engine_matrix[xi][yi]
        for xi,yi in around
    ]    

    return any(map(
        lambda a:a in symbols, adjacency
    ))


print(engine_matrix[139][4])
print(look_around1(139,4), look_around2(139,4), look_around3(139,4))
```

```python
# get coordinates of efery digit grouped by number
num_pos = []
for i in range(len(engine_matrix)):
    number_coords = []
    for j in range(len(engine_matrix)):
        if engine_matrix[i][j] in string.digits:
            number_coords.append((i,j))
        else:
            if len(number_coords) > 0:
                num_pos.append(number_coords)
                number_coords = []
    
    if len(number_coords) > 0:
        num_pos.append(number_coords)
            
num_pos[-1]
```

```python

def has_symbol_adjacent1(pos):
    return any(map(
        lambda p: look_around1(p[0], p[1]), pos
    ))

def has_symbol_adjacent2(pos):
    return any(map(
        lambda p: look_around2(p[0], p[1]), pos
    ))

def has_symbol_adjacent3(pos):
    return any(map(
        lambda p: look_around3(p[0], p[1]), pos
    ))

def pos_to_num(pos):
    num = ""
    for p in pos:
        num += engine_matrix[p[0]][ p[1]]
        
    return int(num)


values1 = [
    pos_to_num(pos)
    for pos in num_pos
    if has_symbol_adjacent1(pos)
]

values2 = [
    pos_to_num(pos)
    for pos in num_pos
    if has_symbol_adjacent2(pos)
]

values3 = [
    pos_to_num(pos)
    for pos in num_pos
    if has_symbol_adjacent3(pos)
]


print("fiirst value:" ,sum(values1), sum(values2), sum(values3))
```

```python

def adjacent_numbers(x,y):
    # just to remain in the matrix (assuming engine_matrix is square)
    mm = lambda c : min(max(c,0), len(engine_matrix)-1)

    around = {
        (mm(x-1),mm(y-1)),
        (x,mm(y-1)),
        (mm(x+1),mm(y-1)),
        (mm(x-1),y),
        (mm(x+1),y),
        (mm(x-1),mm(y+1)),
        (x,mm(y+1)),
        (mm(x+1),mm(y+1))
    }

    adjacent_num = []
    for num in num_pos:
        for pos in num:
            if pos in around:
                adjacent_num.append(num)
                break
    return adjacent_num

def get_gear_ratio(x,y):
    adjacent_num = adjacent_numbers(x,y)
    if len(adjacent_num) != 2:
        return 0
    
    an0 = pos_to_num(adjacent_num[0])
    an1 = pos_to_num(adjacent_num[1])

    return an0 * an1

ratios = []
for i in range(len(engine_matrix)):
    for j in range(len(engine_matrix)):
        if engine_matrix[i][j] == "*":
            ratios.append(get_gear_ratio(i,j))

print("second value: ", sum(ratios))
```
