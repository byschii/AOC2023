```python
card_order = ["A", "K", "Q", "T", "9", "8", "7", "6", "5", "4", "3", "2", "J",]
card_strength = lambda x: len(card_order) - card_order.index(x)
```

```python
# import Counter from collections
from collections import Counter

def hand_strength(hand_str):
    # j is jolly
    hand_str = hand_str.replace("J", "")
    # print(f"hand_str = {hand_str}")
    num_jolly = 5 - len(hand_str)
    hand_counter = Counter(list(hand_str))

    if num_jolly == 5:
        return 7

    # get most common card frequency
    most_common_card_freq = hand_counter.most_common(1)[0][1] + num_jolly

    if most_common_card_freq == 5:
        return 7
    
    # get second most common card frequency
    second_most_common_card_freq = hand_counter.most_common(2)[1][1]
    if most_common_card_freq == 4:
        return 6
    elif most_common_card_freq == 3 and second_most_common_card_freq == 2:
        return 5
    elif most_common_card_freq == 3:
        return 4
    elif most_common_card_freq == 2 and second_most_common_card_freq == 2:
        return 3
    elif most_common_card_freq == 2:
        return 2
    else:
        return 1

# if first < second return -1 else return 1
# if first == second return 0
def first_hand_better(first_hand, second_hand):
    first_hand_strength = hand_strength(first_hand)
    second_hand_strength = hand_strength(second_hand)
    if first_hand_strength > second_hand_strength:
        return 1
    elif first_hand_strength < second_hand_strength:
        return -1
    
    # if both hands have the same strength
    # we need to compare the cards
    for first_card, second_card in zip(first_hand, second_hand):
        first_card_strength = card_strength(first_card)
        second_card_strength = card_strength(second_card)
        if first_card_strength > second_card_strength:
            return 1
        elif first_card_strength < second_card_strength:
            return -1

    return 0   
    
```
