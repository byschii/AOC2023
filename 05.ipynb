The rest of the almanac contains a list of maps which describe how to convert numbers from a source category into numbers in a destination category. That is, the section that starts with seed-to-soil map: describes how to convert a seed number (the source) to a soil number (the destination). This lets the gardener and his team know which soil to use with which seeds, which water to use with which fertilizer, and so on.

Rather than list every source number and its corresponding destination number one by one, the maps describe entire ranges of numbers that can be converted. Each line within a map contains three numbers: the destination range start, the source range start, and the range length.

Consider again the example seed-to-soil map:

50 98 2

With this information, you know that seed number 98 corresponds to soil number 50 and that seed number 99 corresponds to soil number 51.

```python
puzzle_parts = puzzle.split("\n\n")
seeds = [int(x) for x in puzzle_parts[0].split(": ")[1].split(" ")]
puzzle_maps = puzzle_parts[1:]
```

```python
from functools import lru_cache

@lru_cache(maxsize=1_300_000)
def go_throught_map(map_str, enter_value):
    map_lines = map_str.split("\n")[1:]
    for line in map_lines:
        destination, origin, qt_range = [int(x) for x in line.split(" ")]
        if origin <= enter_value <= origin + qt_range:
            return destination + (enter_value - origin)
        
    return enter_value
        
        
for s in seeds:
    print("Seed: ", s, end=" ")
    for p in puzzle_maps:
        s = go_throught_map(p, s)
    print("Result: ", s)        

    
```

```python
@lru_cache(maxsize=900_000)
def seed_to_location(s):
    for p in puzzle_maps:
        s = go_throught_map(p, s)

    return s
```

```python
import time 

i = 0
min_locations = -1
begin = time.time()
for sb, sr in zip(seed_begin, seed_range):
    for s in range(sb, sb + sr):
        s = seed_to_location(s)
        if min_locations == -1 or s < min_locations:
            print("Min Location: ", s, (time.time() - begin )//60)
            min_locations = s
        i += 1
        if i % 100_000 == 0:
            print(i, (time.time() - begin )//60)

print("Min Location: ", min(min_locations))
```
